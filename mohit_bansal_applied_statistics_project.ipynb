### 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

#### 2A. Probability that none of the items are defective?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

n = 10
p = 0.05
k = np.array([0,1,2])

binomial = stats.binom.pmf(k,n,p)
binomial

print("Probablity that none items are defective is :- ", binomial[0])
```

#### 2B. Probability that exactly one of the items is defective? 

```python
print("Probablity that exactly one item is defective is:- ", binomial[1])
```

#### 3B. Probability that two or fewer of the items are defective?

```python
print("Probability that two or fewer items are defective is:- ", binomial[0] + binomial[1] + binomial[2])
```

#### 3C. Probability that three or more of the items are defective ?

```python
Y = 1 - (binomial[0] + binomial[1] + binomial[2])
print("Probability that three or more items are defective is:- ", Y)
```

### 3. A car salesman sells on an average 3 cars per week.

#### 3A. What is Probability that in a given week he will sell some cars?

```python
import numpy as np
import pandas as pd
from scipy.stats import poisson
from scipy.stats import poisson
from matplotlib import pylab as plt

instances = np.array([0,1,2,3,4])     
rate = 3
Poissons = poisson.pmf(instances, rate);
Poissons
prob_one_some_car_sales = 1 - Poissons[0]
print("Probability that salesman will sell some cars is :-", prob_one_some_car_sales)
```

#### 3B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?

```python
prob_between_two_and_four = Poissons[2] + Poissons[3]+ Poissons[4]
print("Probability that between 2 and less than 5 cars are sold :-", prob_between_two_and_four)
```

#### 3C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

```python
x = np.arange(0,10,1)
Poissons = poisson.pmf(x, rate)
Poissons
plt.plot(x, Poissons)
plt.title(f'Poisson: $\lambda$ = {rate}')
plt.xlabel("No of Instances of Cars Sold per Week")
plt.ylabel("Probabililty of no of cars Sold per Week")
plt.show()
```
